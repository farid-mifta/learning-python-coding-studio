```python
# Numerik tipe data
# int -> bil bulat
usia_siswa = 22 # tipe data int (bil bulat)
print(type(usia_siswa)) # cek tipe data
print(usia_siswa) # tampilkan nilai/data

# float -> bil desimal
tinggi_siswa = 171.5 # tipe data float (bil desimal)
print(tinggi_siswa)
print(type(tinggi_siswa))
```

```python
# boolean tipe -> true, false
suka_python = False
print(type(suka_python))
```

```python
# Sequence tipe data

# Tipe data list
# menggunakan []
nama_siswa = ["hafiz", "riana", "isma", "farid"]
print(nama_siswa[0]) # akses index
print(nama_siswa)
print(type(nama_siswa))

# tipe data tuple -> ()
nama_hari = ("senin", "selasa", "rabu")
print(type(nama_hari))

# tipe data range() -> biasa digunakan dalam looping
angka = range(1, 10, 2) # (start, stop, step)
print(type(angka))
```

```python
# set tipe data
# menggunakan {} -> kurung kurawal
himp_siswa = {"isma", "farid", "trisno", "hafiz", "riana"}
print(himp_siswa)
print(type(himp_siswa))
# intersection, join, selisih.
```

```python
# mapping type -> key:value
# tipe dictionary
kontak_telepon = {
    # key:value
    "riana": 998,
    "isma": 997,
    "farid": 996
}

print(kontak_telepon)
print(type(kontak_telepon))
# akses nilai
print(kontak_telepon["isma"])
```

```python
# operator aritmetika
angka_pertama = 10
angka_kedua = 5

print(f"Penjumlahan + : {angka_pertama + angka_kedua}")
# print("Penjumlahan + :", angka_pertama + angka_kedua)
# print(f"penjumlahan {angka_kedua} + {angka_pertama}: {angka_pertama + angka_kedua}")
print(f"Pengurangan - : {angka_pertama - angka_kedua}")
print(f"Pembagian / : {angka_pertama / angka_kedua}")
print(f"Perkalian X : {angka_pertama * angka_kedua}")
print(f"Sisa Hasil Bagi (modulus) %  : {angka_pertama % angka_kedua}")
print(f"pangkat (**) : {angka_pertama ** angka_kedua}")
```

```python
# operator comparison
a = 10
b = 5

print("a > b: ", a > b) # True
print("a < b: ", a < b) # False
print("a >= b: ", a >= b) # True
print("a <= b: ", a <= b) # False
print("a = b: ", a == b) # False
print("a != b: ", a != b) # True
```

```python
# Kontrol flow if
"""
if kondisi:
  blok kode yang akan dieksekusi
"""
nilai_siswa = 90

if nilai_siswa >= 90:
  print("Grade A")
```

```python
# kontrol flow if else
"""
if kondisi:
  blok kode
else:
  blok kode
"""
nilai_siswa = 85

if nilai_siswa >= 90:
  print("grade A")
else:
  print("Kurang Gacorr")
```

```python
# kontrol flow if elif else
nilai_siswa = 75

if nilai_siswa >= 90:
  print("A")
elif nilai_siswa >= 80:
  print("B")
elif nilai_siswa >= 70:
  print("C")
else:
  print("D")
```
